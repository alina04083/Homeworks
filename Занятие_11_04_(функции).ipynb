```python
from math import sqrt
def Mean(x,y):
  A=(x+y)/2
  G=sqrt(x*y)
  return A,G
```

```python
a=2
b=3
c=9
d=77
x1, x2 = Mean(a,b)
y1, y2 = Mean(a,c)
z1, z2 = Mean(a,d)
print(x1, x2, y1, y2, z1, z2)
```

```python
def Minmax(x,y):
  if x>y:
    mx=x
    mn=y
  else:
    mn=x
    mx=y
  return mn,mx
```

```python
a=6
b=3
c=7
d=3
a1, b1 = Minmax(a,b)
c1, d1 = Minmax(c,d)
x, y = Minmax(a1,c1)
x, y = Minmax(b1,d1)
print(x, y)


дорешать
```

```python
def SumRange(A,B):
  z=0
  for i in range (A,B+1):
    if A>B:
      z=0
    else:
      z=z+i
  return z
```

```python
A=int(input('A='))
B=int(input('B='))
C=int(input('C='))
x=SumRange(A,B)
y=SumRange(B,C)
print(x,y)
```
