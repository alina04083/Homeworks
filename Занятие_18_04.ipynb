```python
def print_kth_row(matrix, K):
    if K < 1 or K > len(matrix):
        print("  K")
        return

    kth_row = matrix[K-1]
    print(" K- :", kth_row)
```

```python
matrix = [
    [6, 4, 8],
    [7, 6, 5],
    [9, 1, 4],
    [1, 3, 9],
    [0, 4, 6]
]

K = int(input('K = '))
print_kth_row(matrix, K)
```

```python
def sum_of_rows(matrix):
    if not matrix:
        print(" ")
        return

    row_sums = [sum(row) for row in matrix]
    return row_sums
```

```python
matrix = [
    [6, 4, 8],
    [7, 6, 5],
    [9, 1, 4],
    [1, 3, 9],
    [0, 4, 6]
]

result = sum_of_rows(matrix)
print("   :", result)
```

```python
def count_elements_above_average(matrix):
    M = len(matrix) #  
    N = len(matrix[0]) #  
    counts = []
    for j in range(N):
        column = [matrix[i][j] for i in range(M)]
        average = sum(column) / M
        count = sum(1 for x in column if x > average)
        counts.append(count)
    return counts
```

```python
matrix = [
    [6, 4, 8, 6, 2, 7],
    [7, 6, 5, 8, 2, 9],
    [9, 1, 4, 3, 8, 1],
    [1, 3, 9, 4, 2, 7],
    [0, 4, 6, 2, 8, 5]
]
result = count_elements_above_average(matrix)
print(" ,      :", result)
```

```python
def swap_min_max_in_columns(matrix):
    M = len(matrix)
    N = len(matrix[0])
    for j in range(N):
        column = [matrix[i][j] for i in range(M)]
        min_index = column.index(min(column))
        max_index = column.index(max(column))
        column[min_index], column[max_index] = column[max_index], column[min_index]
        for i in range(M):
            matrix[i][j] = column[i]
    return matrix
```

```python
matrix = [
    [6, 4, 8, 6, 1, 2],
    [7, 6, 5, 8, 2, 9],
    [9, 1, 4, 3, 6, 1],
    [1, 3, 9, 4, 3, 7],
    [0, 4, 6, 2, 8, 5]
]

result = swap_min_max_in_columns(matrix)
print(" :")
for row in result:
    print(row)
```
