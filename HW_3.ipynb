```python
from math import pow, factorial
x=float(input('x='))
n=int(input('n='))
s=0
z=1
for i in range (1, n+1):
  j=2*i-1
  s=s+(pow(x,j)/factorial(j))*z
  z=-z
print(s)
```

```python
from math import factorial
e = 10**(-3)
s = 0
n = int(input('n='))
while True:
    a = 10**n / factorial(n)
    s += a
    n += 1
    if abs(a) < e:
        break
r=round(s, 3)
print(r)
```
